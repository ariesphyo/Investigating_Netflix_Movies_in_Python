```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Subset the dataframe keeping only movies and filtering the released_year in the 1990s
netflix_df_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000) & (netflix_df['type'] == "Movie")]

print(netflix_df_1990s.value_counts('type'))
```

```python
#Change to a new dataframe using pd.DataFrame
movies_1990s = pd.DataFrame(netflix_df_1990s)
movies_1990s.head()
```

```python
#plotting data to view the distributions of durations for movies released in the 1990s

plt.hist(movies_1990s["duration"])
plt.title('Distributions of durations for movies released in the 1990s')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
#create a new dataframe again to keep only movies with the correct genre "Action"

action_movies_1990s = movies_1990s[movies_1990s["genre"] == "Action"]
print(action_movies_1990s.value_counts('genre'))
```

```python
#Iterate through the data and count the number of movies which duration is lesss than 90

short_movie_count = 0

for label, row in action_movies_1990s.iterrows() :
    if row["duration"] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count

print(short_movie_count)
```
